```python
import mne
import mne_microstates
import os
from mne.datasets import sample
```

```python
clean_folder = "/content/drive/My Drive/drdo/subjects"

fname= "/content/drive/My Drive/drdo/channel_loc.csv"
# Read a montage from a file.
TUB_montage = mne.channels.read_custom_montage(fname)

read_data(clean_folder)
```

```python
def read_data(subjectPath):
    
    try:
        root, dirs, files = next(os.walk(clean_folder))

        
        for folder_name in dirs:
          path_file_fif1 = subjectPath + '/' + folder_name + '/nback1.vhdr'
          microstate_analysis(path_file_fif1, folder_name,'nback1')

          path_file_fif2 = subjectPath + '/' + folder_name + '/nback2.vhdr'
          microstate_analysis(path_file_fif1, folder_name,'nback2')

          path_file_fif3 = subjectPath + '/' + folder_name + '/nback3.vhdr'
          microstate_analysis(path_file_fif1, folder_name,'nback3')
            
    except StopIteration:
        pass
        print("Error ocurred:")
        print("Directory with dataset does not found!")
        print("Program will be terminated")
        exit(1)
```

```python
def microstate_analysis(file_path, subject_name, sesssion_name):

  raw = mne.io.read_raw_brainvision(file_path, eog=('HEOG', 'VEOG'), preload=True)  
  raw.set_montage(TUB_montage)

  raw.pick_types(meg='mag', eeg=True, eog=True, ecg=True)

  # Microstate analysis needs average reference
  raw.set_eeg_reference('average')

  # Filter the data. Make sure slow drifts are eliminated.
  raw.filter(1, 40)

  # Clean EOG with ICA
  ica = mne.preprocessing.ICA(0.99).fit(raw)
  bads_eog, _ = ica.find_bads_eog(raw)
  ica.exclude = bads_eog[:2]
  raw = ica.apply(raw)

  # Select sensor type
  raw.pick_types(meg=False, eeg=True)
  # raw.pick_types(meg='mag', eeg=False)

  # Segment the data into 5 microstates
  maps, segmentation, polarity = mne_microstates.segment(raw.get_data(), n_states=5,
                                                         random_state=0,
                                                         return_polarity=True)
    
  # Plot the topographic maps of the microstates and part of the segmentation
  mne_microstates.plot_maps(maps, raw.info)
  mne_microstates.plot_segmentation(segmentation[:500], raw.get_data()[:, :500],
                                    raw.times[:500], polarity=polarity[:500])
```
